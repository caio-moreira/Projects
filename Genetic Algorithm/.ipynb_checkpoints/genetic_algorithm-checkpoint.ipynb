```python
import matplotlib.pyplot as plt
import main

k = 5

bests, means, pop_ini = main.main(k)
```

```python
for i in range(0, k):
    plt.figure(figsize = [4.8, 3.6])
    plt.plot(list(range(0, len(bests[i]))), bests[i], 'r--', list(range(0, len(means[i]))), means[i], 'b-')
    plt.title('Gráfico ' + str(i + 1) + ' - Resultados da população de ' + str(pop_ini[i]) + ' indivíduos')
    plt.xlabel('Geração')
    plt.ylabel('Fitness')
    plt.legend(['Melhor', 'Média'])
    plt.grid(True)
    plt.savefig(str(str(pop_ini[i]) + '.png'), pad_inches = 0.0)
    plt.show()
```
