```python
import algorithm

k = 6

bests, means, pop_size = algorithm.genetic_algorithm(k)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt

for i in range(0, k):
    plt.figure(figsize = [4.8, 3.6])
    plt.plot(list(range(0, len(bests[i]))), bests[i], 'r--', list(range(0, len(means[i]))), means[i], 'b-')
    plt.title('Figure ' + str(i + 1) + ' - Results from ' + str(pop_size[i]) + ' subject population')
    plt.xlabel('Generation')
    plt.ylabel('Fitness')
    plt.legend(['Best', 'Average'])
    plt.grid(True)
    plt.show()
```
